# Requiring Modules 

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Question No 1

```python
a=float(input("Enter the value of a:"))
b=float(input("Enter the value of b:"))      #Taking the range as input
amp = 1 / abs(b-a)                           #Calculating the amp by the user provided range 
x = np.arange(a,b,0.001)                     #creating array for the x axis of the plot 
y = [amp] * x.shape[0]                       #calculating value of f(x) corresponding to the value of x 
plt.plot(x,y)                                #ploting the graph
plt.ylabel("F(x)")                           #setting title of y axis
plt.xlabel("X")                              #setting title of x axis
plt.title("Universal Random Variable")       #setting title for the y axis 
plt.show()                                   #output the graph
```

# Question No 2

```python
def pdfgaus(x,mean = 0, std = 0.5):   #Getting the value of exponent at every value of x by this function
    return np.exp((-np.power((x - mean),2)/2*std)) /(2* std)

mean=float(input("Enter the mean :"))
variance=float(input("Enter the varience :"))
x = np.arange(-5,5,0.001)             #Creating array for the x axis of the plot 
plt.plot(x,pdfgaus(x,mean,variance))
plt.ylabel("F(x)")
plt.xlabel("X")
plt.title("Gaussian Random Variable")
plt.show()
```

# Question No 3

```python
def pdfofbrv(x,n,p): #Getting the value of combinations at every value of x
    return (math.factorial(n) * np.power(p,x) * np.power(1-p,n-x)) / (math.factorial(x)* math.factorial(n-x))

n=int(input("Enter the number of Events :"))      #taking number of trials as input
p=float(input("Enter the success probability :")) #taking the success probability as input
x=np.arange(0,n,1);                               #Storing a array for the x axix 
y=[]
for i in range(n):
    y.append(pdfofbrv(i,n,p))                     #Storing values corresponding to x in the y array for the y axis 

plt.plot(x,y)
plt.ylabel("F(x)")
plt.xlabel("X")
plt.title("Binomial Random Variable")
plt.show()
```
