```python
import pandas as pd
import numpy as np
```

```python
def actFunc(haha,choose):
    if choose==1:
        return (1/(1+np.exp(haha)))
    elif choose==2:
        return np.max(0,haha)
    elif choose==3:
        return np.tanh(haha)
    elif choose==4:
        temp=np.exp(haha)
        return temp/np.sum(temp,axis=1,keepdims=True)
```

```python
def forwardPass(Data):
    a0=Data;
    z1=np.dot(a0,w1)+b1
    a1=actFunc(z1,3)
    z2=np.dot(a1,w2)+b2
    a2=actFunc(z2,4)
    return a0,z1,a1,z2,a2
    
```

```python
def costFunc(y,t):
    return -(t*np.log(y)+(1-t)*np.log(1-y))
```

```python
def Accuracy(y,TrainLabel):
    maxintraindata=np.argmax(y,axis=1)
    maxinlabel=np.argmax(TrainLabel,axis=1)
    count=0
    for l in range(16000):
        if (maxintraindata[l]==maxinlabel[l]):
            count=count+1
    return (count/16000)*100
    
```

```python
for j in range(int(epochs)):
    for i in range(int(16000/batch)):
        a0,z1,a1,z2,a2=forwardPass(trainData[i*batch:(i+1)*batch,:])
        y=a2
        LabelBatch=trainLabel[i*batch:(i+1)*batch,:]
        del2=y-LabelBatch
        del1=np.dot(del2,w2.T)*(1-pow(actFunc(z1,3),2))
        dcdw2=np.dot(a1.T,del2)
        dcdw1=np.dot(a0.T,del1)
        dcdb2=np.sum(del2,axis=0)
        dcdb1=np.sum(del1,axis=0)
        w1=w1-alpha*dcdw1
        w2=w2-alpha*dcdw2
        b1=b1-alpha*dcdb1
        b2=b2-alpha*dcdb2
    a0,z1,a1,z2,a2=forwardPass(trainData)
    ytrue=a2
    cost=np.sum(costFunc(ytrue,trainLabel))/16000
    print("cost")
    print(cost)
    accuracy=Accuracy(ytrue,trainLabel)
    print("Accuracy")
    print(accuracy)
    
      
        
```
