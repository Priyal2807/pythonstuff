```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(-5,5,0.001);
g=1/(1+np.exp(-(np.power(x,1)))); #Sigmiod function
plt.plot(x,g);
plt.show();
```

```python
x=np.arange(-10,10,0.0001);
for i in np.arange(0,len(x)):  #relu function but the negative axis hasn't been figured out
    if x[i]>0:
        g=x;
    else:
        g=0;
    
plt.plot(x,g);
plt.show();
    
```

```python
x=np.arange(-4,4,0.005);
g=(2/(1+np.exp(-np.power(2*x,1))))-1 #Tanh function
plt.plot(x,g);
plt.show();
```

```python
x=np.arange(-10,10,0.0001);
alpha=0.001;
for i in np.arange(0,len(x)):  #leaky relu
    if x[i]>=0:
        g=x;
    else:
        g=alpha*x;
    
plt.plot(x,g);
plt.show();
```
